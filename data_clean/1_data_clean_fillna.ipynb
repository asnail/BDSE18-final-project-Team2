```python
df
# 透過 locat ObservDay_time altitude_classification 去補當天平均值方式後
# 剩下的都是當天全部都是nan的所以再怎麼用這三個特徵去補都補不起來
# 所以將新增一個欄位將台灣分成幾個區塊
# 1. Keelung Taipei New_TaipeiCity Yilan
# 2. Taoyuan Hsinchu Miaoli
# 3. Taichung Changhua Nantou
# 4. Yunlin Chiayi Tainan
# 5. Kaohsiung Pingtung
# 6. Hualien Taitung
```

```python
df['locate_classification']=np.nan
```

```python
ls1=[['Keelung','Taipei','New_TaipeiCity','Yilan'],['Taoyuan','Hsinchu','Miaoli'],['Taichung', 'Changhua', 'Nantou'],['Yunlin' ,'Chiayi' ,'Tainan'],['Kaohsiung','Pingtung'],['Hualien', 'Taitung']]
```
