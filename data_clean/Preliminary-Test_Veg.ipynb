#Data Cleaning

```python
import numpy as np
import pandas as pd
All_Veg=pd.read_csv('./修改成西元年.csv',encoding='utf-8')

# All_Veg.head(3) #8,11447163
All_Veg['作物代號']
```

```python
#確認缺失值
All_Veg.isnull().sum()#0缺失
```

```python
#查看各欄位平均值.四分位數差等資訊
All_Veg.describe().style.format("{:.0f}")
#平均價.交易量min為-782,-139721???
```

```python
#找出平均價,交易量為負者
negative=All_Veg['交易量']<0
All_Veg[negative].index
All_Veg[negative]
#共104個：[ 818052,  834885,  861550,  871002,  874579,  888533,  893136,893564,  903304,  910566,
          #  ...5094909, 5147291, 5158519, 5171126, 5185191, 5352085, 5384858,5504058, 6156219, 7744865]
```

```python
#以0取代負值者
All_Veg2=All_Veg
All_Veg2.at[All_Veg[negative].index,'交易量']=0
# All_Veg2.loc[7744865,'交易量']
```

```python
negative2=All_Veg2['平均價']<0
All_Veg2[negative2]
# All_Veg2[negative2].index #10687411, 11082471, 11386451
```

```python
#作物代碼表中未曾見到ZZZZZ所以直接drop欄位
All_Veg3=All_Veg2.drop([10687411, 11082471, 11386451])
negative3=All_Veg3['平均價']<0
All_Veg3[negative3]
```

```python
#查看各欄位平均值.四分位數差等資訊
All_Veg3.describe().style.format("{:.0f}")
```

```python
All_Veg3
```

```python
#去除過多為0的欄位
All_Veg4=All_Veg3.replace(0,np.nan).dropna(thresh=7)
All_Veg5=All_Veg4.replace(np.nan,0)
```

```python
All_Veg5[All_Veg5['上價']==0]
```

```python
All_Veg5[All_Veg5['中價']==0]
```

```python
All_Veg5[All_Veg5['下價']==0]
```

```python
##以上中下價取代平均價為0者
replace_column=All_Veg5[All_Veg5['平均價']==0]
replace_column['平均價']=(replace_column['上價']+replace_column['中價']+replace_column['下價'])/3

replace_column
# 上價：以當日該農產品總交易量中最高價格之20%，加權平均計算得之。
# 下價：以當日該農產品總交易量中最低價格之20%，加權平均計算得之。
# 中價：以當日該農產品總交易量中扣除最高最低價格各20%剩餘之60%，加權平均計算得之。
All_Veg5.update(replace_column)
```

```python
All_Veg5[All_Veg5['平均價']==0]
```

```python
##保留交易量為0者（待查）
All_Veg5[All_Veg5['交易量']==0]

replace4836382=All_Veg5[All_Veg5['交易量']==0]
replace4836382['平均價']=(replace4836382['上價']+replace4836382['中價']+replace4836382['下價'])/3
replace4836382
```

```python
#4836382:平均價高於上價 以三價和平均取代
All_Veg5[All_Veg5['平均價']==6763.0]
All_Veg5.update(replace4836382)
```

```python
All_Veg5[All_Veg5['交易量']==0]
```

```python
# All_Veg5['市場代號'].astype(int).unique()
```

```python
#取代０,以避免影響平均值
All_Veg6=All_Veg5.replace(0,np.nan)
All_Veg6.isnull().sum()
```

```python
#以同一城市平均填補同城市資料
fillwithAVG=All_Veg6.groupby(['作物代號','市場代號','交易日期']).mean()
fillwithAVG.head(3)
```

```python
#剩餘缺失值數目
fillwithAVG.isnull().sum()
```

```python
fillwithAVG.describe().style.format("{:.0f}")
```

```python
#以同一城市平均填補同城市資料仍有缺失值,以ffill補值先但仍有缺失
fillwithAVG2=fillwithAVG.fillna(method='ffill')
```

```python
fillwithAVG2.isnull().sum()
```
