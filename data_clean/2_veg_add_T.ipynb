```python
ma1=df1['市場代號']==109
ma2=df1['市場代號']==104
df2=df1[ma1|ma2]
```

```python
ma3=df2['交易日期']>'1999-12-31'
df3=df2[ma3]
```

```python
# di1={104:'Taipei',109:'Taipei'}
df3['市場代號']=df3['市場代號'].replace(109,'Taipei')
df3['市場代號']=df3['市場代號'].replace(104,'Taipei')
```

```python
df3['市場代號'].unique()
```

```python
df3
```

```python
test=df3.iloc[:,[0,4]]
test=test.groupby('交易日期').apply(lambda x:x['中價'].mean())
test
test=pd.DataFrame(test)
test
```

```python
test1=df3.iloc[:,[0,7]]
# test1
test1=test1.groupby('交易日期').apply(lambda x:x['交易量'].sum())
test1
test1=pd.DataFrame(test1)
test1
```

```python
df4=df3.iloc[:,[2,0]]
df4
```

```python
df5=df4.drop_duplicates()
```

```python
df5
```

```python
df6=pd.merge(df5,test1,on=['交易日期'])
# last_df=pd.merge(df,veg,on=['ObservDay_time'],how='outer')
```

```python
df7=pd.merge(df6,test,on=['交易日期'])
```

```python
df8 = df7.sort_values(by = '交易日期')
df8.reset_index(inplace=True,drop=True)
```

```python
df8.columns=['Market','ObservDay_time','TradingVolume','MediumAvgPrice']
df8
```

```python
import numpy as np
df8['T+1']=np.nan
df8['T+3']=np.nan
df8['T+5']=np.nan
df8['T+7']=np.nan
df8
```

```python
df8['ObservDay_time']=pd.to_datetime(df8['ObservDay_time'])
```

```python
df8
```

```python
df8['T+1']=((df8['MediumAvgPrice'].shift(-1)-df8['MediumAvgPrice'])/df8['MediumAvgPrice']).apply(lambda x:format(x,'.3f'))
df8['T+3']=((df8['MediumAvgPrice'].shift(-3)-df8['MediumAvgPrice'])/df8['MediumAvgPrice']).apply(lambda x:format(x,'.3f'))
df8['T+5']=((df8['MediumAvgPrice'].shift(-5)-df8['MediumAvgPrice'])/df8['MediumAvgPrice']).apply(lambda x:format(x,'.3f'))
df8['T+7']=((df8['MediumAvgPrice'].shift(-7)-df8['MediumAvgPrice'])/df8['MediumAvgPrice']).apply(lambda x:format(x,'.3f'))
```
