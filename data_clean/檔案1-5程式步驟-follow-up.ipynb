#Data Cleaning

```python
import numpy as np
import pandas as pd
All_Veg=pd.read_csv('./修改成西元年.csv',encoding='utf-8')
cleaned_Typhoon=pd.read_csv('./5-typhoon_ok加入01.csv',low_memory=False)
Typhoon=pd.read_csv('./new_add_typhoon_0and1.csv',low_memory=False)
cleaned_Typhoon.shape
```

```python
cleaned_Typhoon1=cleaned_Typhoon.drop(columns=['ty_code'])
```

```python
cleaned_Typhoon1.isnull().sum()
```

```python
#去掉列中有比9欄還多的列(自訂5欄,已知3欄,共8欄,留至少有一欄剩下的欄位)
cleaned_Typhoon2=cleaned_Typhoon1.dropna(thresh=9)
```

```python
cleaned_Typhoon2.shape
```

```python
#查看缺失值比率(除以最高缺失數：4505323算百分率)
missing_value_ratio_col=((cleaned_Typhoon2.isnull().sum())/2412414).to_frame(name='missing value ratio')
m2=missing_value_ratio_col[missing_value_ratio_col['missing value ratio'] <0.85 ]
m2.index
```

```python
#刪除缺值率>0.85的欄位後所剩欄位['locate', 'StationCode', 'ObservDay_time', 'StnPres', 'SeaPres','Temperature', 'T Max', 'T Min', 'Td dew point', 'RH', 'RHMin', 'WS',
#'WD', 'WSGust', 'WDGust', 'Precp', 'PrecpHour', 'SunShine','SunShineRate', 'GloblRad', 'VisbMean', 'UVI Max', 'Cloud Amount',
#'ty_code', 'ty_generate', 'sea_warning', 'land_warning', 'ty_strength','ty_path']
cleaned_Typhoon3=cleaned_Typhoon2[['locate', 'StationCode', 'ObservDay_time', 'StnPres', 'Temperature',
       'T Max', 'T Min', 'RH', 'RHMin', 'WS', 'WD', 'WSGust', 'WDGust',
       'Precp', 'ty_generate', 'sea_warning', 'land_warning', 'ty_strength',
       'ty_path']]
cleaned_Typhoon3.head(3)
```

```python
#查看各欄位平均值.四分位數差等資訊
cleaned_Typhoon3.describe().style.format("{:.0f}")
#Outlier:
     #Temp:60
     #StnPres min -2?
     #T Max :60
     #T Min :94
```

```python
#找出Temp60度的欄位並且討論過後改為Nan（11月 60度）
# cleaned_Typhoon3[cleaned_Typhoon3['Temperature']>40]
#2187721
cleaned_Typhoon4=cleaned_Typhoon3
cleaned_Typhoon4.at[2187721	,'Temperature']=np.nan
# cleaned_Typhoon4[cleaned_Typhoon4['Temperature']>40]
cleaned_Typhoon4.describe().style.format("{:.0f}")

#Outlier:
     #StnPres min -2?
     #Tmax=60,Tmin=94,-40
```

```python
#找出Tmax60度的欄位,進一步搜尋T Max>50者
#C0I480:梅峰;C0A650:火燒寮
cleaned_Typhoon4[cleaned_Typhoon4['T Max']>50]
# cleaned_Typhoon4[cleaned_Typhoon3['T Max']>50].index
```

```python
#並且討論過後令[1830549, 2215679, 2215680, 2215683, 2215803, 2262144, 2347452,2347453, 2347454, 2347455]為Nan
cleaned_Typhoon5=cleaned_Typhoon4
cleaned_Typhoon5.at[1830549,'T Max']=np.nan
cleaned_Typhoon5.at[2215679,'T Max']=np.nan
cleaned_Typhoon5.at[2215680,'T Max']=np.nan
cleaned_Typhoon5.at[2215683,'T Max']=np.nan
cleaned_Typhoon5.at[2215803,'T Max']=np.nan
cleaned_Typhoon5.at[2262144,'T Max']=np.nan
cleaned_Typhoon5.at[2347452,'T Max']=np.nan
cleaned_Typhoon5.at[2347453,'T Max']=np.nan
cleaned_Typhoon5.at[2347454,'T Max']=np.nan
cleaned_Typhoon5.at[2347455,'T Max']=np.nan

cleaned_Typhoon5.describe().style.format("{:.0f}")
```

```python
#找出Tmin94度的欄位,進一步搜尋T min>50者:[1479539, 1479540, 1479541, 2264039, 3769139, 3769140]
#C0E550:明德;C0A920:富貴角;C0S760:紅石
cleaned_Typhoon6=cleaned_Typhoon5
cleaned_Typhoon6[cleaned_Typhoon6['T Min']>50]
# cleaned_Typhoon5[cleaned_Typhoon4['T Min']>50].index
```

```python
#以Nan取代Ｔ Min:[1479539, 1479540, 1479541, 2264039, 3769139, 3769140] 
cleaned_Typhoon6.at[1479539,'T Min']=np.nan
cleaned_Typhoon6.at[1479540,'T Min']=np.nan
cleaned_Typhoon6.at[1479541,'T Min']=np.nan
cleaned_Typhoon6.at[2264039,'T Min']=np.nan
cleaned_Typhoon6.at[3769139,'T Min']=np.nan
cleaned_Typhoon6.at[3769140,'T Min']=np.nan

cleaned_Typhoon6.describe().style.format("{:.0f}")
```

```python
#找出StnPres -2
#C0X170:關廟,C0V710:苓雅
cleaned_Typhoon6[cleaned_Typhoon6['StnPres']<500]
#index:[1225248, 1225249, 3413173, 3413174, 3413175, 3413176, 3413177]
```

```python
#苓雅有颱風保留;關廟不太可能氣壓那麼低,以NaN取代
cleaned_Typhoon7=cleaned_Typhoon6
cleaned_Typhoon7.at[1225248,'StnPres']=np.nan
cleaned_Typhoon7.at[1225249,'StnPres']=np.nan

# cleaned_Typhoon7[cleaned_Typhoon7['StnPres']<500]
###<600有：C0I54:奇萊稜線A;C0R240:來義;
cleaned_Typhoon7.describe().style.format("{:.0f}")
```

```python
#以同一城市平均填補同城市資料
fillwithAVG=cleaned_Typhoon7.groupby(['locate','ObservDay_time']).mean()
# fillwithAVG.info()
```

```python
#剩餘缺失值數目
fillwithAVG.isnull().sum()
```

```python
#以同一城市平均填補同城市資料仍有缺失值,以ffill補值先但仍有缺失,改以bfill
fillwithAVG2=fillwithAVG.fillna(method='bfill')
# fillwithAVG2.reset_index().to_csv('filled_clean_ty.csv')
```

```python
#確認無缺失值
fillwithAVG2.isnull().sum()
```
